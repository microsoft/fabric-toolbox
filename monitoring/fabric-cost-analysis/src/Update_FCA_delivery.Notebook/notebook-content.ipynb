```python
import sempy.fabric as fabric
```

```python
def delete_notebook_if_exists(
    notebook_display_name: str,
    workspace: str | None = None,
    dry_run: bool = True
) -> bool:
    """
    If a notebook with the given Display Name exists in the workspace, delete it.
    Returns True if a deletion would happen (or happened), False otherwise.
    """
    items = fabric.list_items(workspace=workspace)

    # Fabric items table typically includes: 'Type', 'Display Name', 'Id'
    matches = items[
        (items["Type"].str.lower() == "notebook") &
        (items["Display Name"] == notebook_display_name)
    ]

    if matches.empty:
        print(f"✅ Notebook not found: {notebook_display_name!r}")
        return False

    if len(matches) > 1:
        raise ValueError(
            f"Multiple notebooks found with name {notebook_display_name!r}. "
            f"Please disambiguate by Id.\n{matches[['Display Name','Id','Type']]}"
        )

    notebook_id = matches.iloc[0]["Id"]
    print(f"🧾 Found notebook: {notebook_display_name!r} (Id={notebook_id})")

    if dry_run:
        print("🟡 Dry run ON — not deleting. Set dry_run=False to delete.")
        return True

    fabric.delete_item(notebook_id, workspace=workspace)
    print("🗑️ Deleted.")
    return True
```

```python
delete_notebook_if_exists(notebook_display_name="01_Load_Calandar", dry_run=False)
```
