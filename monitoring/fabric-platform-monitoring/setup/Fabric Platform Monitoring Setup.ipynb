# Process

## Deployment Logic
This part iterates through all the items, gets the respective source code, replaces all IDs dynamically and deploys the new item

```python
import fabric_deployment_tool
import notebookutils

exclude = []
type_exclude = []

if not INSTALL_ACTIVITY_EVENTS_MODULE:
    type_exclude.append("ActivityEvents")

if not INSTALL_INVENTORY_MODULE:
    type_exclude.append("Inventory")

if not INSTALL_GATEWAY_MODULE:
    type_exclude.append("Gateway")

if not INSTALL_CAPACITY_MODULE:
    type_exclude.append("Capacity")

pipeline_parameters = {
    "key_vault_uri": key_vault_uri,
    "key_vault_tenant_id": key_vault_tenant_id,
    "key_vault_client_id": key_vault_client_id,
    "key_vault_client_secret": key_vault_client_secret,   
}

fabDeploymentTool = fabric_deployment_tool.FabDeploymentTool()
```

```python
fabDeploymentTool.set_github(repo_owner=repo_owner, repo_name=repo_name, branch=branch, folder_prefix=folder_prefix, github_token=github_token)
fabDeploymentTool.run(exclude=exclude, type_exclude=type_exclude, workspace_name=workspace_name, capacity_name=capacity_name, eventhouse_name=eventhouse_name, pipeline_parameters=pipeline_parameters)
```
